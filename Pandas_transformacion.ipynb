### Precision de valores numericos





Por lo general, cuando los datos son muy grandes y tenemos poca memoria disponible, es habitual utilizar tipos de datos más compactos para reducir el consumo de memoria. Sin embargo, siempre es importante asegurarse de que la elección del tipo de datos no perjudique la precisión o exactitud de los resultados.

Cuando trabajamos con números enteros con Python, podemos tener diferentes tipos de datos, cada uno con sus limitaciones y características. Durante las clases trabajamos con int64, un número entero con precisión de 64 bits. Para entender el significado de esta precisión, es importante conocer algunos términos técnicos, como byte y bit:

**Byte:** es una unidad de medida de información, que representa un conjunto de 8 bits.
**Bit:** es la unidad de información más pequeña utilizada en los sistemas digitales, y puede tomar los valores de 0 o 1.
Tipo entero

Con los conceptos de bit y byte claros, podemos comprender mejor el significado de int64, que es el tipo de entero que utiliza 8 bytes de almacenamiento: 8 bits en cada byte, lo que da como resultado 64 bits en total. Este tipo de entero es capaz de representar números muy grandes, que pueden oscilar entre -9.223.372.036.854.775.808 y 9.223.372.036.854.775.807.

Además, tenemos otros números enteros cuya precisión se puede definir, como int32, un tipo de datos entero que utiliza 4 bytes - 8 bits en cada byte, lo que da como resultado 32 bits en total. Es capaz de representar números enteros menores que los representados por int64, con un máximo de -2.147.483.648 a 2.147.483.647.

Puede ser más común encontrar los tipos int64 y int32, pero puede ser necesario, en algunas situaciones, utilizar otros tipos de datos enteros, como int8 o int16. Tipos de datos como estos son útiles cuando necesita ahorrar más memoria y no trabaja con valores grandes. Puede consultar los tipos de números enteros en la siguiente tabla:

Tipo de dado	|Número de bits|	Valor mínimo	| Valor máximo
---|---|---|---
|int8|	8|	-128	| 127
|int16|	16|	-32.768|	32.767
|int32|	32|	-2.147.483.648	|2.147.483.647
|int64	|64	|-9.223.372.036.854.775.808|	9.223.372.036.854.775.807

La elección de qué valor de precisión dependerá de la situación y la naturaleza de los datos que se manipulan. Si los valores que se analizan son relativamente pequeños, el uso de int32, por ejemplo, puede ser suficiente y puede ahorrar espacio en la memoria. Sin embargo, si estuviéramos trabajando con datos científicos, por ejemplo, que requieren valores muy grandes, es posible que necesitemos usar int64.

Tipo float

Además de los números enteros, otros tipos de datos, como float, también utilizan esta opción de precisión como opción para controlar el espacio de memoria. Al igual que los números enteros, el tipo float también tiene opciones de precisión: entre los tipos más comunes se encuentran float32 y float64.

El tipo float64 es un número de punto flotante con 64 bits de precisión, que representa un número decimal de hasta 15 dígitos. Por otro lado, float32 es más pequeño tanto en su capacidad de bits, con 32 en total, como en su capacidad de precisión decimal, con capacidad de precisión de hasta 7 dígitos.

## Tokenizacion


Tokenización: Extrayendo Valor de los Textos

En el contexto de la valoración de alquileres, la tokenización es una técnica clave para analizar datos de texto, como las descripciones de las propiedades. Estas descripciones contienen información valiosa sobre las características y comodidades que influyen en el precio del alquiler.

¿Qué es la Tokenización?

La tokenización es el proceso de dividir un texto en unidades más pequeñas llamadas "tokens". Estos tokens pueden ser palabras, sílabas o incluso conjuntos de palabras, dependiendo de la aplicación. En nuestro caso, realizaremos una tokenización simple, donde cada palabra del texto se convertirá en un token.

¿Por qué Tokenizar?

No podemos sumar, dividir o contar directamente un texto. La tokenización nos permite transformar el texto en unidades que podemos analizar cuantitativamente. Al identificar palabras clave y patrones en las descripciones, podemos determinar qué características tienen mayor impacto en el precio del alquiler.

Ejemplo Práctico

Imagina que analizamos las descripciones de varios alquileres y encontramos que palabras como "excelente", "fantástico" o "maravilloso" aparecen con frecuencia en propiedades de alto precio. Por otro lado, palabras como "horrible" o "no recomendado" podrían estar asociadas con alquileres más económicos.

Pasos para la Tokenización

Normalización del Texto: Antes de tokenizar, es importante normalizar el texto. Esto incluye convertir todo el texto a minúsculas y eliminar caracteres extraños que puedan interferir con el análisis.
División en Tokens: Una vez normalizado el texto, lo dividimos en tokens. En nuestro caso, cada palabra se convierte en un token individual.
Análisis de Tokens: Finalmente, analizamos los tokens para identificar patrones y relaciones con el precio del alquiler. Podemos contar la frecuencia de cada palabra, identificar palabras clave y utilizar técnicas de aprendizaje automático para predecir el precio en función de las palabras presentes en la descripción.

### Normalizar con regex

En la ciencia de datos, las expresiones regulares se utilizan a menudo para procesar datos de texto sin formato. Algunas de estas aplicaciones incluyen limpieza de datos, donde se pueden usar expresiones regulares para buscar y reemplazar ciertos caracteres. También se puede utilizar en proyectos de clasificación; un ejemplo es el uso de expresiones regulares para ayudar a encontrar patrones en el texto de los correos electrónicos, lo que ayuda a clasificar si son spam o no.

En general, las expresiones regulares permiten a los científicos de datos procesar, analizar y clasificar grandes volúmenes de datos de texto de forma eficiente y automatizada. El uso adecuado de expresiones regulares puede ayudar a extraer información valiosa de los datos de texto, además de facilitar la limpieza y organización de esos datos.

Puedes crear una expresión regular con la ayuda del sitio web [regex101.com](https://regex101.com/).
````python
# examino una unica variable para ver que caracteres eliminar
# por ejemplo comas, asteriscos, dos puntos, barras diagonales.
# dejo las comillas de palabas compuestas en ingles o guiones.
datos['descripcion_local'][3169]
````

````python
# pido que reemplace con regex todo lo que NO SEA (^) [^a-zA-Z0-9\'-]
# a esta secuencia se la llama Expresion Regular.
# letras de la 'a' a la 'z' y la 'A' a la 'Z'
# \'- comilla y guion

datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\'-]',' ', regex=True)
````

````python
datos['descripcion_local'][3169]
````

````python
# pido que reemplace el guion si funciona como espacio entre - dos - palabras
# le  digo que reeplace los guiones que no tengan un caracter a la  izquierda o derecha. (?<!\w)
# la r atras de la expresion regular para qeu python la interprete literalmente y no como secuencias de escape.
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r'(?<!\w)-(?<!\w)',' ', regex=True)
datos['descripcion_local'].head()
````

### que es regex?
En la función replace de Pandas, el parámetro regex se utiliza para indicar si la búsqueda y el reemplazo deben interpretarse como una expresión regular (regex).

Cuando regex=True, el patrón de búsqueda se trata como una expresión regular, lo que permite realizar búsquedas más complejas y flexibles. Puedes usar metacaracteres y patrones especiales para encontrar y reemplazar texto de manera más precisa.

Cuando regex=False (o se omite, ya que este es el valor predeterminado), el patrón de búsqueda se trata como una cadena literal. En este caso, la búsqueda se realiza de forma exacta, sin interpretar ningún carácter especial.

### porque uno tiene corchetes?
La diferencia en el uso de corchetes entre las dos expresiones regulares se debe a que tienen propósitos distintos:

Expresión regular con corchetes: [^a-zA-Z0-9\'-]

Los corchetes se utilizan para definir una "clase de caracteres". En este caso, la clase de caracteres incluye todas las letras (mayúsculas y minúsculas), los números del 0 al 9, la comilla simple (') y el guion (-).
El símbolo ^ al principio de la clase indica que se busca cualquier carácter que NO esté dentro de esa clase. Es decir, se busca cualquier carácter que no sea una letra, un número, una comilla simple o un guion.
Esta expresión regular se utiliza para eliminar todos los caracteres especiales que no son letras, números, comillas simples o guiones.
Expresión regular sin corchetes: (\W)-(\W)

En esta expresión regular, no se utilizan corchetes porque no se está definiendo una clase de caracteres. En cambio, se está buscando un patrón específico: un carácter que no sea una palabra (\W), seguido de un guion (-), seguido de otro carácter que no sea una palabra (\W).
Los paréntesis se utilizan para crear "grupos de captura". Estos grupos permiten referirse a las partes coincidentes del patrón en el reemplazo. En este caso, se capturan los caracteres no alfanuméricos antes y después del guion.
Esta expresión regular se utiliza para eliminar los guiones que están rodeados de caracteres no alfanuméricos, es decir, los guiones que no forman parte de palabras compuestas.

## Etapa 4

En las columnas datas_de_pagamento y datas_combinadas_pagamento tenemos fechas en el formato 'día/mes/año' (dd/mm/AAAA). Transforme estos datos al tipo datetime y busque una forma de visualización de subconjunto que pueda contribuir al objetivo del contexto en el que se insertan los datos.

Objetivo: **analizar el retraso en el pago de la renta** en el **condominio** ficticio de algunos **residentes**.

* Agrupo por apartamento (eso agrupa por residente porque no tengo columna residente).
* La renta se paga el dia acordado en data de pagamento.
* puedo hacer un booleano de comparar data de pago y de pagamento
* lo almaceno en una nueva columna
* group by departamento y nueva columna
* la comparacion da cantidad de true pagos a tiempo y cantidad de false pagos retrasados.
* el informe dice porcentaje de alquileres retrasados por apartamento. ordenados del mas moroso almenos. se informa la cantidad de meses de alquiler por contrato para brindar mas informacion.

````python
# Tranformacion a datetime
datos_administracion_condominios['datas_de_pagamento'] = pd.to_datetime(datos_administracion_condominios['datas_de_pagamento'], format='%d/%m/%Y')
````

````python
datos_administracion_condominios['datas_combinadas_pagamento'] = pd.to_datetime(datos_administracion_condominios['datas_combinadas_pagamento'], format='%d/%m/%Y')
````

````python
#  COLUMNA BOOLEANA
# si  pago a tiempo  ('datas_de_pagamento')<= (datas_combinadas_pagamento) --> True
#dia = datos_administracion_condominios['datas_de_pagamento'].dt.day

datos_administracion_condominios['pago a tiempo'] = datos_administracion_condominios['datas_de_pagamento'] <= datos_administracion_condominios['datas_combinadas_pagamento']
datos_administracion_condominios
````

````python
datos_administracion_condominios['retraso'] = datos_administracion_condominios['datas_de_pagamento'] - datos_administracion_condominios['datas_combinadas_pagamento']
datos_administracion_condominios
````

````python
apartamentos = datos_administracion_condominios[['apartamento','pago a tiempo','retraso']].sort_values(by= 'apartamento')
media_retraso = apartamentos.groupby(['apartamento'])['retraso'].mean().sort_values(ascending=False)
media_retraso
````
