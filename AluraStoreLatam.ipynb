## Productos de todas las tiendas que mas ingresos generaron


```python
# CONCATENACION DE DF DE TIENDAS
# Creo una lista [] de urls sobre la que iterar
urls = ["https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
"https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
"https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
"https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"]

#pd.read_csv() lee directamente la URL .assign agrega la columna tienda
df_total = pd.concat(
                    [pd.read_csv(url).assign(Tienda = f'Tienda {i+1}') for i, url in enumerate(urls)],
                     ignore_index=True
                     )
df_total
```

```python
#PREPARACION DE DATOS
#Buena practica: convertir a Datetime la columna de fecha

df_total['Fecha de Compra'] = pd.to_datetime(df_total['Fecha de Compra'], dayfirst=True)
df_total.sample(15)

```

```python
from numpy import dtype
# PREPARACION DE DATOS
# Renombramos y tipamos de ser necesario nuestras columnas.
#tipamos precio para que sea de tipo float

df_total['Precio'] = df_total['Precio'].astype(float)
dtype(df_total['Precio'])
```

```python
# CALCULO DE INGRESOS Y VENTAS AGRUPADAS
# Calculo Ingreso total del producto en cada tienda
ventas = df_total.groupby(['Tienda', 'Producto'])['Precio'].sum().reset_index(name='Ingresos_Totales')
ventas
```

```python
from IPython.core import formatters
# Creo una funcion de analisis de tiendas

def analizar_tienda(df_tienda, nombre_tienda):

  #ordeno mis ingresos totales
  ordenado = df_tienda.sort_values(by='Ingresos_Totales', ascending=False)

  # Top 3 productos que generaron mas ingresos
  print(f'\n {nombre_tienda}: Top 3 productos que mas ingresos generaron')
  print(ordenado.head(3).to_string(
      index=False, columns=['Producto', 'Ingresos_Totales'],
      formatters={'Ingresos_Totales': '${:,.2f}'.format}
  ))

  # Top 3 productos que generaron menos ingresos
  print(f'\n {nombre_tienda}: Top 3 productos que menos ingresos generaron')
  print(ordenado.tail(3).to_string(
      index=False, columns=['Producto', 'Ingresos_Totales'],
      formatters={'Ingresos_Totales': '${:,.2f}'.format}
  ))
```

```python
# ITERAR Y ANALIZAR
#Creo un bucle que itera sobre el nombre de cada tienda en mi array creado con unique.
for nombre_tienda in ventas['Tienda'].unique():
  #creo un sata frame por tienda comparando si es igual al nomrbede la tienda sobre la que itera mi bucle.
  data_tienda = ventas[ventas['Tienda'] == nombre_tienda]
  #uso el data frame por tienda de parametro para la funcion
  analizar_tienda(data_tienda, nombre_tienda)
```

### Facturacion acumulada por tienda, trimestral.

```python
df_mensual = (
    df_total.set_index('Fecha de Compra')
            .groupby('Tienda')['Precio']
            .resample('Q')
            .sum()
            .unstack(0)
)

df_acum = df_mensual.cumsum()
df_acum.plot(figsize=(12,6))
plt.title('Facturación acumulada por tienda')
plt.ylabel('Ingresos acumulados')
plt.show()
```

### Calificación promedio por tienda, trimestral.

```python
df_mensual = df_total.set_index('Fecha de Compra').groupby('Tienda')['Calificación'].resample('Q').mean().unstack(0)

df_mensual.plot(kind='line', figsize=(12,6))

plt.title('Evolución de la calificación promedio por tienda')
plt.xlabel('Fecha')
plt.ylabel('Calificación promedio')
plt.legend(title='Tienda')
plt.grid(True)
plt.show()
```

### Facturación trimestral por tienda (no acumulada)

```python
df_mensual = df_total.set_index('Fecha de Compra').groupby('Tienda')['Precio'].resample('Q').sum().unstack(0)

df_mensual.plot(kind='line', figsize=(12,6))

plt.title('Evolución de los ingresos totales por tienda')
plt.xlabel('Fecha')
plt.ylabel('Ingresos totales')
plt.legend(title='Tienda')
plt.grid(True)
plt.show()
```
