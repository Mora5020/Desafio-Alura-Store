# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.


```python
def ventas(tienda, numero):
  vpc_tienda = tienda['Categoría del Producto'].value_counts()
  print(f'La cantidad de ventas por categoria de la tienda {numero} es: \n {vpc_tienda}')

ventas(tienda, 1)
ventas(tienda2, 2)
ventas(tienda3,  3)
ventas(tienda4, 4)


```

### Facturacion por categoria


```python

def venta_por_categoria(tienda, numero):
  ventas_por_categoria = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
  print(f'Las ventas por categoria de la tienda {numero} son \n{ventas_por_categoria}')

venta_por_categoria(tienda, 1)
venta_por_categoria(tienda2, 2)
venta_por_categoria(tienda3, 3)
venta_por_categoria(tienda4, 4)
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
#Funcion para calcular el promedio
def calcular_promedio(tienda,numero):
  promedio = tienda['Calificación'].mean()
  print(f'La calificacion promedio de la Tienda {numero}: {promedio:.2f}')
  #{promedio:.2f} indica la cantidad de decimales del resultado

#calcular promedio de cada tienda
calcular_promedio(tienda, 1)
calcular_promedio(tienda2, 2)
calcular_promedio(tienda3, 3)
calcular_promedio(tienda4, 4)
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

## Productos mas y menos vendidos de cada tienda

```python
def maximos_minimos_vendidos(tienda, numero):
  ventas = tienda['Producto'].value_counts()
  mas_vendidos = ventas.head(10)
  menos_vendidos = ventas.tail(10)
  menos_vendidos.sort_values(ascending=True, inplace=True)
  print(f'Los productos mas vendidos de tienda {numero} son: \n{mas_vendidos}\n\nLos productos menos vendidos de la tienda {numero} son: \n{menos_vendidos}')

maximos_minimos_vendidos(tienda, 1)
maximos_minimos_vendidos(tienda2, 2)
maximos_minimos_vendidos(tienda3, 3)
maximos_minimos_vendidos(tienda4, 4)
```

## Productos de todas las tiendas que mas ingresos generaron


```python
# CONCATENACION DE DF DE TIENDAS
# Creo una lista [] de urls sobre la que iterar
urls = ["https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
"https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
"https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
"https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"]

#pd.read_csv() lee directamente la URL .assign agrega la columna tienda
df_total = pd.concat(
                    [pd.read_csv(url).assign(Tienda = f'Tienda {i+1}') for i, url in enumerate(urls)],
                     ignore_index=True
                     )
df_total
```

```python
df_total.columns
```

```python
#PREPARACION DE DATOS
#Buena practica: convertir a Datetime la columna de fecha

df_total['Fecha de Compra'] = pd.to_datetime(df_total['Fecha de Compra'], dayfirst=True)
df_total.sample(15)

```

```python
from numpy import dtype
# PREPARACION DE DATOS
# Renombramos y tipamos de ser necesario nuestras columnas.
#tipamos precio para que sea de tipo float

df_total['Precio'] = df_total['Precio'].astype(float)
dtype(df_total['Precio'])
```

```python
# CALCULO DE INGRESOS Y VENTAS AGRUPADAS
# Calculo Ingreso total del producto en cada tienda
ventas = df_total.groupby(['Tienda', 'Producto'])['Precio'].sum().reset_index(name='Ingresos_Totales')
ventas
```

```python
from IPython.core import formatters
# Creo una funcion de analisis de tiendas

def analizar_tienda(df_tienda, nombre_tienda):

  #ordeno mis ingresos totales
  ordenado = df_tienda.sort_values(by='Ingresos_Totales', ascending=False)

  # Top 3 productos que generaron mas ingresos
  print(f'\n {nombre_tienda}: Top 3 productos que mas ingresos generaron')
  print(ordenado.head(3).to_string(
      index=False, columns=['Producto', 'Ingresos_Totales'],
      formatters={'Ingresos_Totales': '${:,.2f}'.format}
  ))

  # Top 3 productos que generaron menos ingresos
  print(f'\n {nombre_tienda}: Top 3 productos que menos ingresos generaron')
  print(ordenado.tail(3).to_string(
      index=False, columns=['Producto', 'Ingresos_Totales'],
      formatters={'Ingresos_Totales': '${:,.2f}'.format}
  ))
```

```python
# ITERAR Y ANALIZAR
#Creo un bucle que itera sobre el nombre de cada tienda en mi array creado con unique.
for nombre_tienda in ventas['Tienda'].unique():
  #creo un sata frame por tienda comparando si es igual al nomrbede la tienda sobre la que itera mi bucle.
  data_tienda = ventas[ventas['Tienda'] == nombre_tienda]
  #uso el data frame por tienda de parametro para la funcion
  analizar_tienda(data_tienda, nombre_tienda)
```

```python
def grafico_ingresos_totales():
  tiendas = [ingreso_total, ingreso_total2, ingreso_total3, ingreso_total4]
  etiquetas_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']


  plt.figure(figsize=(6,6))

  plt.pie(
          tiendas,
          labels=etiquetas_tiendas,
          autopct='%1.1f%%',
          startangle=90
      )
  plt.title('Aporte de cada tienda a los ingresos totales')
  plt.axis('equal')

  plt.show()

grafico_ingresos_totales()
```

### Facturacion acumulada por tienda, trimestral.

```python
def facturacion_en_tiempo():
  df_mensual = (
    df_total.set_index('Fecha de Compra')
            .groupby('Tienda')['Precio']
            .resample('Q')
            .sum()
            .unstack(0)
  )

  df_acum = df_mensual.cumsum()
  df_acum.plot(figsize=(12,6))
  plt.title('Facturación acumulada por tienda')
  plt.ylabel('Ingresos acumulados')
  plt.show()

facturacion_en_tiempo()
```

### Calificación promedio por tienda, trimestral.

```python
def calificacion_trimestral():
  df_mensual = df_total.set_index('Fecha de Compra').groupby('Tienda')['Calificación'].resample('Q').mean().unstack(0)

  df_mensual.plot(kind='line', figsize=(12,6))

  plt.title('Evolución de la calificación promedio por tienda')
  plt.xlabel('Fecha')
  plt.ylabel('Calificación promedio')
  plt.legend(title='Tienda')
  plt.grid(True)
  plt.show()
calificacion_trimestral()
```

### Facturación trimestral por tienda (no acumulada)

```python
def facturacion_trimestral():

  df_mensual = df_total.set_index('Fecha de Compra').groupby('Tienda')['Precio'].resample('Q').sum().unstack(0)

  df_mensual.plot(kind='line', figsize=(12,6))

  plt.title('Facturacion trimestral por tienda')
  plt.xlabel('Fecha')
  plt.ylabel('Ingresos totales')
  plt.legend(title='Tienda')
  plt.grid(True)
  plt.show()
facturacion_trimestral()
```

## Facturacion por categoria

```python
def categorias_mas_vendidas():
  ventas_por_categoria = df_total.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum().unstack()

  mas_vendidas = ventas_por_categoria.head()

  ax = mas_vendidas.plot(kind='bar', figsize=(12,8), grid=True)
  ax.set_ylabel('Facturacion en $')
  ax.set_xlabel('Tiendas')
  ax.set_title('Categorias mas vendidas en cada tienda')
  ax.ticklabel_format(style='plain', axis='y')

  print(categorias_mas_vendidas)


categorias_mas_vendidas()
```

```python
def categorias_menos_vendidas():
  ventas_categoria = df_total.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum().unstack()

  menos_vendidas = ventas_categoria.head()

  ax = menos_vendidas.plot(kind='bar', figsize=(12,8), grid=True)
  ax.set_ylabel('Facturacion en $')
  ax.set_xlabel('Tiendas')
  ax.set_title('Categorias menos vendidas en cada tienda')
  ax.ticklabel_format(style='plain', axis='y')

  print(categorias_menos_vendidas)


categorias_menos_vendidas()
```
